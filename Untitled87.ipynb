```python
# Import necessary libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Create the dataset
data = {
    "area": [2600, 3000, 3200, 3600, 4000],
    "price": [550000, 565000, 610000, 680000, 725000]
}
df = pd.DataFrame(data)

# Prepare the data
X = df[['area']]  # Feature (2D array)
y = df['price']   # Target (1D array)

# Create and train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predict prices for areas in the dataset
predicted_prices = model.predict(X)
df['predicted_price'] = predicted_prices

# Display the dataset with actual and predicted prices
print("Dataset with Actual and Predicted Prices:")
print(df)

# Plot the data and regression line
plt.scatter(df['area'], df['price'], color='blue', label='Actual Prices')
plt.plot(df['area'], predicted_prices, color='red', label='Regression Line')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price')
plt.legend()
plt.show()
```
