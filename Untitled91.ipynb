```python
# Import necessary libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df = pd.read_csv('cars.csv')  # Ensure 'cars.csv' is in the same directory as this script

# Prepare the data
X = df[['Cars (sq ft)']]  # Feature (independent variable)
y = df['Price ($)']       # Target (dependent variable)

# Create and train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predict prices for the cars in the dataset
predicted_prices = model.predict(X)

# Plot the data and regression line
plt.scatter(df['Cars (sq ft)'], df['Price ($)'], color='blue', label='Actual Prices')  # Actual data points

# Create a wider range for the regression line
extended_cars = np.linspace(min(df['Cars (sq ft)']), max(df['Cars (sq ft)']), 100).reshape(-1, 1)
predicted_line = model.predict(extended_cars)
plt.plot(extended_cars, predicted_line, color='red', label='Regression Line')  # Regression line

# Chart labels and legend
plt.xlabel('Cars (sq ft)')
plt.ylabel('Price ($)')
plt.title('Car Price Prediction Based on Cars (sq ft)')
plt.legend()

# Show the plot
plt.show()
```
