```python
# Import necessary libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

# Create the dataset
data = {
    "area": [2600, 3000, 3200, 3600, 4000],
    "price": [550000, 565000, 610000, 680000, 725000]
}
df = pd.DataFrame(data)

# Prepare the data
X = df[['area']]  # Feature (2D array)
y = df['price']   # Target (1D array)

# Create and train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predict prices for areas 5000, 8000, and 9000 sq ft
areas_to_predict = [[5000], [8000], [9000]]
predicted_prices = model.predict(areas_to_predict)

# Display predictions
for area, price in zip([5000, 8000, 9000], predicted_prices):
    print(f"The predicted price for a home with area {area} sq ft is {price:.2f}")

# Plot the data and regression line
plt.scatter(df['area'], df['price'], color='blue', label='Actual Prices')  # Actual data points

# Create a wider range for the regression line
extended_areas = np.linspace(2600, 9000, 100).reshape(-1, 1)  # From 2600 to 9000 sq ft
predicted_line = model.predict(extended_areas)
plt.plot(extended_areas, predicted_line, color='red', label='Regression Line')  # Extended regression line

# Add predicted points
predicted_areas = [5000, 8000, 9000]
plt.scatter(predicted_areas, predicted_prices, color='green', label='Predicted Prices')  # Predicted points

# Chart labels and legend
plt.xlabel('Area (sq ft)')
plt.ylabel('Price')
plt.title('Home Price Prediction')
plt.legend()

# Show the plot
plt.show()
```
